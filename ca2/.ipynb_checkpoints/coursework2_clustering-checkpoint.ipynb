```python
#import libraries
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
import numpy as np

from statistics import mean
from copy import deepcopy
from scipy.spatial import distance
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
def get_data(col_names, filename):
    """
    Parameters:
        col_names: the names to assign to the columns
        filename: the name of the file to read in
    Returns:
        data: the data stored within the file
    """
    data = pd.read_csv(filename,
                    sep=',', # In the file, attributes are separated by white spaces
                    names=col_names)
    return data
```

# WP1

```python
def kmeans(x, k, max_itr=100):
    """
    Parameters:
        x: the data do be clustered
        k: the number of clusters
        max_itr: the maximum number of iterations
    Returns:
        cluster_labels: the cluster membership labels for each element in the data x
    """
    features = x.columns[:]
    labels = x[x.columns[-1]].unique()
    num_features = len(features)
    
    Centroids = (x.sample(n=k))
    difference = 1
    first_run = True
    
    while(difference !=0 and max_itr > 0):
        
        # find the distance to each cluster
        x_differences=x
        i=1
        for centroid_row in Centroids.itertuples():
            distances = []
            a = centroid_row[1:num_features+1]
            for data_row in x_differences.itertuples():
                b = data_row[1:num_features+1]
                distances.append(distance.euclidean(a, b))
            x[i]=distances
            i=i+1

        # assign the closest cluster's index to each data point
        cluster=[]
        for index,row in x.iterrows():
            min_dist=row[1]
            pos=1
            for i in range(k):
                if row[i+1] < min_dist:
                    min_dist = row[i+1]
                    pos=i+1
            cluster.append(pos)
        x["Cluster"]=cluster
        
        # check for 0 difference
        Centroids_new = x.groupby(["Cluster"]).mean()[features]
        if first_run:
            difference=1
            first_run = False
        else:
            # sum the difference between all new and old centroids
            difference = 0
            for feature in features:
                difference += (Centroids_new[feature] - Centroids[feature]).sum()

        Centroids = x.groupby(["Cluster"]).mean()[features]
        max_itr -= 1

    return Centroids.values, x.iloc[:,-1].values
```

## WP2

```python
def average_distance(centroid, cluster):
    """
    Parameters:
        centroid: The coordinates of the clusters centroid
        cluster: The coordinates of each point in the cluster
    Returns:
        The average distance between each point in the cluster and the centroid
    """
    distance_sum = 0
    for point in cluster:
        distance_sum += distance.euclidean(centroid, point)
    return distance_sum / len(cluster)

def compute_all_Dij(cluster_i, cluster_j, centroid_i, centroid_j):
    """
    Parameters:
        cluster_i: The coordinates of a point in the ith cluster
        cluster_j: The coordinates of a point in the jth cluster
        centroid_i: The coordinates of the centroid in the ith cluster
        centroid_j: The coordinates of the centroid in the jth cluster
    Returns:
        The average intra cluster distance between clusters i and j
    """
    dij = distance.euclidean(centroid_i, centroid_j)
    return (average_distance(centroid_i, cluster_i) + average_distance(centroid_j, cluster_j)) / dij

def get_cluster(target_cluster, x, cluster_labels):
    """
    Parameters:
        target_cluster: the index of the cluster to find
        x: the data that has been clustered
        cluster_labels: the cluster membership labels for each element in the data x
    Returns:
        cluster: The coordinates of each point in the target cluster
    """
    cluster = []
    for i in range(len(cluster_labels)):
        if cluster_labels[i] == target_cluster+1:
            cluster.append(x.iloc[i])
    return cluster
    
def davies_bouldin(x, cluster_labels, centroids, k):
    """
    Parameters:
        x: the data that has been clustered
        cluster_labels: the cluster membership labels for each element in the data x
        centroids: the co-ordinates of the centroids for the data in x
    Returns:
        The Davies-Bouldin index for the given data
    """
    davies_bouldin_max = []
    for i in range(k):
        davies_bouldin_curr = []
        for j in range(k):
            if i != j:
                cluster_i = get_cluster(i, x, cluster_labels)
                cluster_j = get_cluster(j, x, cluster_labels)
                davies_bouldin_curr.append(compute_all_Dij(cluster_i, cluster_j, centroids[i], centroids[j]))
        davies_bouldin_max.append(max(davies_bouldin_curr))
    return sum(davies_bouldin_max) / k
```

## WP3

```python
def silhouette_scores(x, cluster_labels):
    """
    Parameters:
        x: the data to be clustered
        cluster_labels: the cluser membership vector produced by the k-means algorithm
    Returns:
        scores: a vector containing the silhouette score for each data sample in x
    """
    ai_values = []
    bi_values = []
    cluster_sizes = []
    
    # find the size of each cluster
    for i in range(1, max(cluster_labels)+1):
        cluster_sizes.append(np.count_nonzero(cluster_labels == i))
    
    # find ai and bi
    for row_i, clusters_i in zip(x.iterrows(), cluster_labels):
        i_cluster_total = 0
        bi_dict = {}
        for row_j, clusters_j in zip(x.iterrows(), cluster_labels):
            if row_i[0] != row_j[0]:
                if clusters_i == clusters_j: # find ai
                    i_cluster_total += distance.euclidean(row_i[1], row_j[1])
                else: # find bi
                    bi_dict.setdefault(clusters_j, 0)
                    bi_dict[clusters_j] += distance.euclidean(row_i[1], row_j[1])
        
        bi_list = []
        for key in bi_dict:
            bi_list.append(bi_dict.get(key)/cluster_sizes[key-1])
        bi_values.append(min(bi_list))
          
        ai_values.append(i_cluster_total / (cluster_sizes[clusters_i-1] - 1))
    
    # use ai and bi to find silhouette score
    si_values = []
    for ai, bi in zip(ai_values, bi_values):
        si_values.append(((bi - ai) / max(ai, bi)))

    return si_values
```

## WP4

```python
def plot_indexes(data, title, max_k=10, n=5):
    """
    Parameters:
        data: The data to be clustered
        title: The title of the graph
        max_k: The maximum number of clusters
    """
    db_indexes_averages = []

    for k in range(2, max_k+1):
        db_indexes_all = []
        for i in range(n):
            x = deepcopy(data)
            centroids, labels = kmeans(x, k)
            db_indexes_all.append(davies_bouldin(data, labels, centroids, k))
        db_indexes_averages.append(mean(db_indexes_all))
    
    plt.figure(figsize=(14,8))
    plt.plot(range(2, max_k+1), db_indexes_averages, 'ro--') # Plot the noise level against the accuracy
    plt.ylabel('Davies-Bouldin values')
    plt.xlabel('Number of clusters (k)')
    plt.title('Davies-Bouldin index for: ' + title)
    plt.show()
```

## The cluster-validation set

```python
data = get_data(['feature1', 'feature2'], "cluster_validation_data.txt")
plot_indexes(data, "cluster_validation_data.txt")
```

The Davies-Bouldin index for the cluster_validation_data shows us that the optimal number of clusters is 3, with 2 being a close second. All other number of clusters have a significantly higher Davies Bouldin index.

## WP5

```python
def plot_silhouette(X,max_k):
    """
    Parameters:
        X: the data to be clustered
        max_k: the maximum number of clusters
    """
    range_n_clusters = range(2,max_k+1)

    for n_clusters in range_n_clusters:
        x1 = deepcopy(X)
        fig, (ax1, ax2) = plt.subplots(1, 2)
        fig.set_size_inches(18, 7)

        # The 1st subplot is the silhouette plot
        ax1.set_xlim([-0.1, 1])
        ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

        centroids, cluster_labels = kmeans(x1,n_clusters)
        sample_silhouette_values = np.asarray(silhouette_scores(X, cluster_labels), dtype=np.float32)
        silhouette_avg = np.mean(sample_silhouette_values)

        y_lower = 10
        for i in range(n_clusters):
            # Aggregate the silhouette scores for samples belonging to cluster i, and sort them
            ith_cluster_silhouette_values = \
                sample_silhouette_values[cluster_labels == i+1]

            ith_cluster_silhouette_values.sort()

            size_cluster_i = ith_cluster_silhouette_values.shape[0]
            y_upper = y_lower + size_cluster_i

            color = cm.nipy_spectral(float(i) / n_clusters)
            ax1.fill_betweenx(np.arange(y_lower, y_upper),
                              0, ith_cluster_silhouette_values,
                              facecolor=color, edgecolor=color, alpha=0.7)

            # Label the silhouette plots with their cluster numbers at the middle
            ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

            # Compute the new y_lower for next plot
            y_lower = y_upper + 10  # 10 for the 0 samples

        ax1.set_title("The silhouette plot for the various clusters.")
        ax1.set_xlabel("The silhouette coefficient values")
        ax1.set_ylabel("Cluster label")

        # The vertical line for average silhouette score of all the values
        ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

        ax1.set_yticks([])  # Clear the yaxis labels / ticks
        ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

        # 2nd Plot showing the actual clusters formed
        colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
        ax2.scatter(X.iloc[:, 0], X.iloc[:, 1], marker='.', s=30, lw=0, alpha=0.7,
                    c=colors, edgecolor='k')

        # Draw white circles at cluster centers
        ax2.scatter(centroids[:, 0], centroids[:, 1], marker='o',
                    c="white", alpha=1, s=200, edgecolor='k')

        for i, c in enumerate(centroids):
            ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1,
                        s=50, edgecolor='k')

        ax2.set_title("The visualization of the clustered data.")
        ax2.set_xlabel("Feature space for the 1st feature")
        ax2.set_ylabel("Feature space for the 2nd feature")

        plt.suptitle(("Silhouette analysis for KMeans clustering "
                      "with n_clusters = %d" % n_clusters),
                     fontsize=14, fontweight='bold')
```

A value of k=4 shows a detoriation in clustering quality as two of the partition's maximums do not exceed the mean. The partitions are also not equally sized and the mean has decreased to below 0.6.
